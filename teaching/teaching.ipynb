```python
list1 = []
list2 = []

list_of_rows = contents.split("\n")
print("Num Rows:", len(list_of_rows))

for row in list_of_rows: # an element is a row in the text file (i.e. row 0 = '77442   88154')
    split_row = row.split("   ") # an element is split into two strings (i.e. the first string value in element 0 is '77442' and the second is '88154')
    list1.append(int(split_row[0]))
    list2.append(int(split_row[1]))
print(list1)
print(list2)

print(sorted(list1))
print(sorted(list2))
```
